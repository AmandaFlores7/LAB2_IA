An√°lisis exploratorio de los datos, que muestren las relaciones que existen entre los atributos y la variable que se busca predecir

```python
# Lectura de datos de un archivo .csv
# Se importa la libreria pandas y se le asigna el alias pd
import pandas as pd
# Se lee el archivo y se asigna a la variable redWines y whiteWines
redWines = pd.read_csv('winequality-red.csv', sep=';')
whiteWines = pd.read_csv('winequality-white.csv', sep=';')

print(redWines.head(1))
```

```python
#Analizar el dataframe con los estadisticos descriptivos de cada variable
redWines.describe().T
```

```python
#Analizar el dataframe con los estadisticos descriptivos de cada variable
whiteWines.describe().T
```

```python
#Informacion relevante del dataset redWines
redWines.info()
```

```python
#Informacion relevante del dataset whiteWines
whiteWines.info()
```

Todos los datos son num√©ricos, por lo que no es necesario realizar un an√°lisis de los datos categ√≥ricos.

```python
#Valores de la variable quality del dataset redWines
redWines['quality'].value_counts()

```

```python
#Valores de la variable quality del dataset whiteWines
whiteWines['quality'].value_counts()
```

Se puede observar que la escala de los datos para ambos van de 3 al 8 para redWine y 3 al 9 para whiteWine

```python
#Histograma de la variable quality en ambos datasets (redWines y whiteWines)
import matplotlib.pyplot as plt
plt.hist(redWines['quality'], 10, facecolor='red', alpha=0.5, label='Red wines')
plt.hist(whiteWines['quality'], 10, facecolor='grey', alpha=0.5, label='White wines')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.title('Histogram of Quality')
plt.legend()
plt.show()

```

```python
#Informacion relevante del dataset redWines en relacion a la variable quality (media, mediana, desviacion estandar, minimo, maximo)
redWines['quality'].describe()
```

```python
#Informacion relevante del dataset whiteWines en relacion a la variable quality (media, mediana, desviacion estandar, minimo, maximo)
whiteWines['quality'].describe()
```

```python
#Correlacion entre las variables del dataset redWines a trav√©s de un mapa de calor
import seaborn as sns
sns.heatmap(redWines.corr(), annot=True, cmap='RdYlGn', linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```

```python
#Correlacion entre las variables del dataset redWines a trav√©s de un mapa de calor en relacion a la variable quality 
import seaborn as sns
sns.heatmap(redWines.corr()[['quality']], annot=True, cmap='RdYlGn', linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```

Analisando los atributos relacionados con la calidad en redWine, se puede observar que los 3 atributos con mayor correlaci√≥n son alcohol, sulphates y volatile acidity.
